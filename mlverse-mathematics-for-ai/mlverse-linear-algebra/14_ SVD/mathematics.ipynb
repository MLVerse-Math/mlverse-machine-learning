# Singular Value Decomposition (SVD) Mathematics

Singular Value Decomposition (SVD) is one of the most important matrix factorization techniques in Linear Algebra, Machine Learning, and Artificial Intelligence.

For any matrix $A \in \mathbb{R}^{m \times n}$, there exists a decomposition:

$$
A = U \Sigma V^T
$$

where:

* $U$ = Left Singular Vector Matrix
* $\Sigma$ = Singular Value Matrix
* $V^T$ = Right Singular Vector Matrix

---

## Orthogonality Properties

The columns of $U$ are orthonormal:

$$
U^T U = I
$$

The columns of $V$ are orthonormal:

$$
V^T V = I
$$

For singular vectors:

$$
u_i^T \nu_j = 0 \quad (i \neq j)
$$

$$
v_i^T v_j = 0 \quad (i \neq j)
$$

This means the singular vectors are mutually orthogonal and represent independent directions in space.

---

## Singular Value Matrix

The matrix $\Sigma$ is diagonal:

$$
\Sigma =
\begin{bmatrix}
\sigma_1 & 0 & 0 & \cdots \
0 & \sigma_2 & 0 & \cdots \
0 & 0 & \sigma_3 & \cdots \
\vdots & \vdots & \vdots & \ddots
\end{bmatrix}
$$

with:

$$
\sigma_1 \geq \sigma_2 \geq \sigma_3 \geq \cdots \geq 0
$$

The singular values measure the importance of each direction in the data.

---

## Connection with Eigenvalues

Consider the matrix:

$$
A^T A
$$

This matrix is always:

* Symmetric
* Positive Semi-Definite

Therefore it has real eigenvalues.

The eigenvalue equation is:

$$
A^T A v_i = \lambda_i v_i
$$

The eigenvectors $v_i$ form the matrix $V$.

The singular values are computed as:

$$
\sigma_i = \sqrt{\lambda_i}
$$

Thus singular values are the square roots of the eigenvalues of $A^T A$.

---

## Right Singular Vectors

The eigenvectors of:

$$
A^T A
$$

form:

$$
V =
\begin{bmatrix}
v_1 & v_2 & \cdots & v_n
\end{bmatrix}
$$

These vectors describe important directions in the input space.

---

## Left Singular Vectors

The left singular vectors are computed as:

$$
u_i = \frac{A v_i}{\sigma_i}
$$

These vectors form the matrix:

$$
U =
\begin{bmatrix}
u_1 & u_2 & \cdots & u_m
\end{bmatrix}
$$

---

## Geometric Interpretation

SVD decomposes a matrix transformation into three operations:

1. Rotation
2. Scaling
3. Rotation

Mathematically:

$$
A = U \Sigma V^T
$$

The matrix $V^T$ rotates the coordinate system.

The matrix $\Sigma$ scales the space.

The matrix $U$ performs a final rotation.

---

## Unit Circle Interpretation

Consider a unit circle.

After applying matrix $A$:

$$
x' = Ax
$$

the circle becomes an ellipse.

The principal axes of the ellipse are determined by the singular vectors.

The lengths of the axes are determined by the singular values.

---

## Rank and SVD

The rank of a matrix equals the number of non-zero singular values.

$$
Rank(A)
=======

#(\sigma_i > 0)
$$

Example:

$$
\Sigma =
\begin{bmatrix}
10 & 0 & 0 \
0 & 4 & 0 \
0 & 0 & 0
\end{bmatrix}
$$

Since two singular values are non-zero:

$$
Rank(A)=2
$$

---

## Low-Rank Approximation

A matrix can be approximated using only the largest singular values.

$$
A_k
===

U_k \Sigma_k V_k^T
$$

where:

* $U_k$ contains the first $k$ left singular vectors
* $\Sigma_k$ contains the first $k$ singular values
* $V_k$ contains the first $k$ right singular vectors

This reduces storage while preserving most information.

---

## Eckart–Young Theorem

The optimal rank-$k$ approximation of a matrix is:

$$
A_k
===

U_k \Sigma_k V_k^T
$$

No other rank-$k$ matrix can achieve lower reconstruction error.

This theorem forms the mathematical foundation of:

* PCA
* Image Compression
* Recommendation Systems
* Embedding Compression

---

## Reconstruction Error

The approximation error is measured using the Frobenius norm:

$$
||A - A_k||_F
$$

The error equals:

$$
\sqrt{
\sum_{i=k+1}^{r}
\sigma_i^2
}
$$

where $r$ is the rank of the matrix.

---

## Image Compression

An image can be represented as a matrix:

$$
A \in \mathbb{R}^{m \times n}
$$

Using SVD:

$$
A = U \Sigma V^T
$$

Keeping only the largest singular values:

$$
A_k = U_k \Sigma_k V_k^T
$$

provides:

* Smaller storage
* Noise reduction
* Minimal visual quality loss

---

## PCA and SVD

Principal Component Analysis computes eigenvectors of:

$$
A^T A
$$

Since SVD already computes these directions, PCA is commonly implemented using SVD.

The principal components correspond to directions with the largest singular values.

---

## Recommendation Systems

User-item interaction matrices can be factorized as:

$$
R = U \Sigma V^T
$$

This reveals hidden latent factors representing:

* User preferences
* Product characteristics

This idea powers systems such as Netflix, Amazon, and Spotify.

---

## Deep Learning Applications

SVD is widely used for:

### Weight Matrix Compression

$$
W = U \Sigma V^T
$$

### Model Optimization

Analyzing:

* Neural network layers
* Embedding matrices
* Attention weights

### Large Language Models

Low-rank approximations derived from SVD are commonly used for efficient training and inference.

---

## Key Takeaways

* Every matrix can be decomposed as:

$$
A = U \Sigma V^T
$$

* Singular values are:

$$
\sigma_i = \sqrt{\lambda_i}
$$

* Rank equals the number of non-zero singular values.

* SVD provides the best low-rank approximation.

* PCA is built on SVD.

* SVD powers recommendation systems, computer vision, NLP, embeddings, deep learning, and modern AI systems.
