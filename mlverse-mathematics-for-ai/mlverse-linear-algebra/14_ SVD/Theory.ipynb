# Singular Value Decomposition Theory

## What is SVD?

SVD decomposes any matrix into three matrices:

$$
A = U\Sigma V^T
$$

where:

- U = Left Singular Vectors
- Σ = Singular Values
- Vᵀ = Right Singular Vectors

---

## Why is SVD Special?

Unlike Eigenvalue Decomposition:

- Works for rectangular matrices
- Always exists
- Numerically stable

---

## Geometric Interpretation

SVD performs three operations:

1. Rotate
2. Scale
3. Rotate Again

Data transformation becomes easy to understand.

---

## Singular Values

Singular values indicate the importance of each direction in the data.

Large singular values:

- Carry useful information

Small singular values:

- Often represent noise

---

## Rank and SVD

Number of non-zero singular values equals matrix rank.
