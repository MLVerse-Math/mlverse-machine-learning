```python
import numpy as np

np.set_printoptions(
    precision=4,
    suppress=True
)

print("NumPy Version:", np.__version__)
```

# Creating a Matrix

We start with a simple matrix:

$$
A=
\begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
$$

```python
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])

print(A)
```

# Computing SVD

SVD factorizes a matrix as:

$$
A = U \Sigma V^T
$$

```python
U, S, VT = np.linalg.svd(A)

print("U Matrix:\n")
print(U)

print("\nSingular Values:\n")
print(S)

print("\nV^T Matrix:\n")
print(VT)
```

# Understanding U Matrix

U contains Left Singular Vectors.

Properties:

$$
U^TU = I
$$

```python
print(
    np.round(
        U.T @ U,
        4
    )
)
```

# Understanding Singular Values

Singular values indicate how important each direction is.

Larger values contain more information.

```python
for i, value in enumerate(S):

    print(
        f"σ{i+1} = {value:.4f}"
    )
```

# Understanding Vᵀ

Rows of Vᵀ are Right Singular Vectors.

Properties:

$$
V^TV = I
$$

```python
print(
    np.round(
        VT @ VT.T,
        4
    )
)
```

# Building Sigma Matrix

NumPy returns singular values separately.

We convert them into a diagonal matrix.

```python
Sigma = np.zeros(
    (A.shape[0], A.shape[1])
)

np.fill_diagonal(
    Sigma,
    S
)

print(Sigma)
```

# Matrix Reconstruction

Using:

$$
A = U\Sigma V^T
$$

we should recover the original matrix.

```python
A_reconstructed = (
    U @ Sigma @ VT
)

print(
    np.round(
        A_reconstructed,
        4
    )
)
```

# Reconstruction Error

Difference between original and reconstructed matrix.

```python
error = np.linalg.norm(
    A - A_reconstructed
)

print(
    "Error:",
    error
)
```

# Rank from SVD

Rank equals the number of non-zero singular values.

```python
rank = np.sum(
    S > 1e-10
)

print(
    "Rank:",
    rank
)
```

# Low Rank Approximation

Keep only the largest singular value.

```python
k = 1

U_k = U[:, :k]

S_k = np.diag(
    S[:k]
)

VT_k = VT[:k, :]

A_k = (
    U_k
    @
    S_k
    @
    VT_k
)

print(
    np.round(
        A_k,
        4
    )
)
```

# Information Retained

Measure how much information remains.

```python
retained = (
    np.sum(S[:1])
    /
    np.sum(S)
)

print(
    f"{retained*100:.2f}%"
)
```

# Condition Number

Measures numerical stability.

$$
\kappa=
\frac{\sigma_{max}}
{\sigma_{min}}
$$

```python
condition_number = (
    S.max()
    /
    S.min()
)

print(
    condition_number
)
```

# Noise Reduction Idea

Small singular values often correspond to noise.

```python
A = np.random.rand(
    10,
    10
)

U,S,VT = np.linalg.svd(A)

S_filtered = S.copy()

S_filtered[5:] = 0

Sigma_filtered = np.zeros(
    (10,10)
)

np.fill_diagonal(
    Sigma_filtered,
    S_filtered
)

A_denoised = (
    U
    @
    Sigma_filtered
    @
    VT
)

print(
    A_denoised.shape
)
```

# Connection with Eigenvalues

For:

$$
A^TA
$$

Singular values are:

$$
\sigma_i=\sqrt{\lambda_i}
$$

```python
A = np.array([
    [1,2],
    [3,4]
])

ATA = A.T @ A

eigenvalues, _ = np.linalg.eig(
    ATA
)

print(
    "Eigenvalues:\n",
    eigenvalues
)

print(
    "\nSquare Root:\n",
    np.sqrt(eigenvalues)
)
```

# PCA using SVD

```python
from sklearn.decomposition import PCA

data = np.random.randn(
    100,
    5
)

pca = PCA()

pca.fit(data)

print(
    pca.explained_variance_
)
```
