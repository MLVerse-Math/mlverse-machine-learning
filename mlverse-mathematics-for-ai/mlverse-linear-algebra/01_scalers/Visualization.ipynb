```python
import matplotlib.pyplot as plt

scalar = 5

plt.figure(figsize=(8,2))

plt.axhline(0)

plt.scatter(scalar,0,s=200)

plt.title("Scalar on Number Line")

plt.yticks([])

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

values = [-4,-2,0,3,6]

plt.figure(figsize=(8,2))

for v in values:
    plt.scatter(v,0,s=150)

plt.axhline(0)

plt.title("Positive and Negative Scalars")

plt.yticks([])

plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1,2])

k = 3

scaled = k*v

plt.figure(figsize=(6,6))

plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1)
plt.quiver(0,0,scaled[0],scaled[1],angles='xy',scale_units='xy',scale=1)

plt.xlim(0,4)
plt.ylim(0,7)

plt.grid(True)

plt.title("Scalar Multiplication")

plt.show()
```

```python
import matplotlib.pyplot as plt

metrics = ["Loss","Accuracy","LR"]
values = [0.25,0.96,0.001]

plt.figure(figsize=(6,4))

plt.bar(metrics,values)

plt.title("AI Metrics Represented as Scalars")

plt.show()
```
