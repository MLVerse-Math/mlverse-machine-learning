```python
# Creating Scalars

a = 10
b = 3.14

print("Integer Scalar:", a)
print("Float Scalar:", b)
```

```python
# Scalar Arithmetic

a = 20
b = 5

print("Addition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a/b)
```

```python
# Absolute Value

a = -15

print(abs(a))
```

```python
# Scalar Multiplication

import numpy as np

v = np.array([1,2,3])

k = 4

print(k*v)
```

```python
# Scalars in Machine Learning

learning_rate = 0.001
loss = 0.25
accuracy = 0.96

print("Learning Rate:", learning_rate)
print("Loss:", loss)
print("Accuracy:", accuracy)
```

```python
# Scalar Types

integer_scalar = 10
float_scalar = 5.6
complex_scalar = 3 + 2j

print(integer_scalar)
print(float_scalar)
print(complex_scalar)
```
