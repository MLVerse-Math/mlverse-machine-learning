# Mathematics of Eigenvalues and Eigenvectors

## Eigenvalue Equation
$$
Av=λv
$$

---

## Rearranging

$$
Av-\lambda v=0
$$

$$
(A-\lambda I)v=0
$$

---

## Characteristic Equation

Non-trivial solutions exist when:
$$

\det(A−λI)=0
$$

---

## Characteristic Polynomial

Example:

$$
A=
\begin{bmatrix}
2&1\\
1&2
\end{bmatrix}
$$

$$
\det(A-\lambda I)
=
(2-\lambda)^2-1
$$

---

## Solving

$$
\lambda^2-4\lambda+3=0
$$

$$
(\lambda-3)(\lambda-1)=0
$$

Eigenvalues:

$$
\lambda_1=3
$$

$$
\lambda_2=1
$$

---

## Finding Eigenvectors

Substitute eigenvalues into:

$$
(A-\lambda I)v=0
$$

Solve for v.

---

## Properties

### Trace

$$
\sum \lambda_i = Tr(A)
$$

### Determinant

$$
\prod \lambda_i = det(A)
$$

### Orthogonal Matrices

Eigenvectors often form orthogonal bases.
