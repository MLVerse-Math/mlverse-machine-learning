# Eigenvalues and Eigenvectors Theory

## Motivation

Consider a matrix transformation.

Most vectors change:

- Direction
- Magnitude

Some special vectors keep their direction.

Only their length changes.

These vectors are Eigenvectors.

---

## Eigenvector Definition

A non-zero vector v is an eigenvector of matrix A if:
$$
Av=λv
$$

where:

- A = Matrix
- v = Eigenvector
- λ = Eigenvalue

---

## Interpretation

The matrix stretches or compresses the vector.

Direction remains unchanged.

---

## Why Important?

Eigenvectors reveal:

- Dominant directions
- Data structure
- Variance directions
- Stable system states

---

## Geometric Meaning

Eigenvectors represent invariant directions under transformation.
