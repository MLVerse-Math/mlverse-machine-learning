# 🎬 Interactive Linear Transformation Explorer

A Linear Transformation maps vectors from one vector space to another.

Mathematically:

$$
T(x)=Ax
$$

where:

$$
A=
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
$$

Use the sliders below to change the matrix entries and observe how the vector transforms in real time.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact
from ipywidgets import FloatSlider

plt.rcParams["figure.figsize"] = (8,8)

vector = np.array([2,1])

def transformation_explorer(a,b,c,d):

    A = np.array([
        [a,b],
        [c,d]
    ])

    transformed = A @ vector

    plt.figure()

    plt.quiver(
        0,0,
        vector[0],
        vector[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        0,0,
        transformed[0],
        transformed[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.grid(True)

    plt.title(
        f"Transformation Matrix\n{A}"
    )

    plt.show()

interact(
    transformation_explorer,

    a=FloatSlider(
        value=1,
        min=-3,
        max=3,
        step=0.1
    ),

    b=FloatSlider(
        value=0,
        min=-3,
        max=3,
        step=0.1
    ),

    c=FloatSlider(
        value=0,
        min=-3,
        max=3,
        step=0.1
    ),

    d=FloatSlider(
        value=1,
        min=-3,
        max=3,
        step=0.1
    )
);
```

# 🎬 Scaling Transformation Explorer

Scaling changes the size of vectors.

Scaling Matrix:

$$
S=
\begin{bmatrix}
s_x & 0\\
0 & s_y
\end{bmatrix}
$$

Move the sliders to stretch or compress the vector.

```python
from ipywidgets import interact

vector = np.array([2,3])

def scaling_animation(sx,sy):

    S = np.array([
        [sx,0],
        [0,sy]
    ])

    result = S @ vector

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,
        vector[0],
        vector[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        0,0,
        result[0],
        result[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.xlim(-15,15)
    plt.ylim(-15,15)

    plt.grid(True)

    plt.title(
        f"Scaling ({sx:.1f},{sy:.1f})"
    )

    plt.show()

interact(
    scaling_animation,

    sx=(-5,5,0.1),

    sy=(-5,5,0.1)
);
```

# 🎬 Rotation Animation

Rotation rotates vectors around the origin.

Rotation Matrix:

$$
R=
\begin{bmatrix}
\cos\theta & -\sin\theta\\
\sin\theta & \cos\theta
\end{bmatrix}
$$

Move the slider to rotate continuously.

```python
vector = np.array([4,2])

def rotation_animation(theta):

    angle = np.radians(theta)

    R = np.array([
        [np.cos(angle),-np.sin(angle)],
        [np.sin(angle), np.cos(angle)]
    ])

    result = R @ vector

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,
        vector[0],
        vector[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        0,0,
        result[0],
        result[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.xlim(-6,6)
    plt.ylim(-6,6)

    plt.grid(True)

    plt.title(
        f"Rotation = {theta}°"
    )

    plt.show()

interact(
    rotation_animation,
    theta=(0,360,1)
);
```

# 🎬 Shear Transformation Animation

Shearing slants shapes while preserving parallel lines.

Matrix:

$$
\begin{bmatrix}
1 & k\\
0 & 1
\end{bmatrix}
$$

Adjust the shear factor.

```python
square = np.array([
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]
])

def shear_animation(k):

    A = np.array([
        [1,k],
        [0,1]
    ])

    transformed = square @ A.T

    plt.figure(figsize=(8,8))

    plt.plot(
        square[:,0],
        square[:,1],
        linewidth=3,
        label="Original"
    )

    plt.plot(
        transformed[:,0],
        transformed[:,1],
        linewidth=3,
        label="Transformed"
    )

    plt.xlim(-5,5)
    plt.ylim(-5,5)

    plt.grid(True)

    plt.legend()

    plt.title(
        f"Shear Factor = {k:.2f}"
    )

    plt.show()

interact(
    shear_animation,
    k=(-3,3,0.05)
);
```

# 🎬 Grid Transformation Playground

Observe how an entire coordinate system transforms.

This provides the geometric intuition behind:

- Neural Networks
- PCA
- Transformers
- Computer Vision

```python
def grid_transform(a,b,c,d):

    A = np.array([
        [a,b],
        [c,d]
    ])

    plt.figure(figsize=(9,9))

    for x in np.arange(-5,6):

        line = np.array([
            [x,-5],
            [x,5]
        ])

        transformed = line @ A.T

        plt.plot(
            transformed[:,0],
            transformed[:,1]
        )

    for y in np.arange(-5,6):

        line = np.array([
            [-5,y],
            [5,y]
        ])

        transformed = line @ A.T

        plt.plot(
            transformed[:,0],
            transformed[:,1]
        )

    plt.xlim(-15,15)
    plt.ylim(-15,15)

    plt.grid(True)

    plt.title(
        f"Matrix\n{A}"
    )

    plt.show()

interact(
    grid_transform,

    a=(-3,3,0.1),

    b=(-3,3,0.1),

    c=(-3,3,0.1),

    d=(-3,3,0.1)
);
```

# 🤖 AI Connection

Neural networks repeatedly apply transformations:

$$
y = Wx + b
$$

where:

- \(x\) = input vector
- \(W\) = transformation matrix
- \(b\) = bias vector

Every hidden layer creates a new transformed representation of the data.

This is why understanding Linear Transformations is essential for:

- Deep Learning
- Computer Vision
- Reinforcement Learning
- Transformers
- Large Language Models
