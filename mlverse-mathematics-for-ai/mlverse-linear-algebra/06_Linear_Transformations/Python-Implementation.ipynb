# 1️⃣ Creating Transformation Matrices

A Linear Transformation can be represented as:

$$
T(x)=Ax
$$

where:

- A is a matrix
- x is a vector

The matrix determines how the vector changes.

```python
A = np.array([
    [2,0],
    [0,2]
])

print("Transformation Matrix:")
print(A)
```

# 2️⃣ Matrix-Vector Multiplication

Applying a transformation means multiplying:

$$
Ax
$$

where:

$$
x=
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$

```python
A = np.array([
    [2,0],
    [0,2]
])

x = np.array([
    3,
    4
])

result = A @ x

print("Original Vector:")
print(x)

print("\nTransformed Vector:")
print(result)
```

# 3️⃣ Scaling Transformation

Scaling changes the size of a vector.

Scaling Matrix:
$$
\begin{bmatrix}
s_x&0\\
0&s_y
\end{bmatrix}
$$

```python
scale = np.array([
    [3,0],
    [0,2]
])

vector = np.array([2,3])

scaled_vector = scale @ vector

print("Original:", vector)
print("Scaled:", scaled_vector)
```

# 4️⃣ Rotation Transformation

Rotation rotates vectors around the origin.

Rotation Matrix:

$$
\begin{bmatrix}
\cos\theta&-\sin\theta\\
\sin\theta&\cos\theta
\end{bmatrix}
$$

```python
theta = np.radians(45)

rotation = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

vector = np.array([1,0])

rotated = rotation @ vector

print("Original:", vector)
print("Rotated :", rotated)
```

# 5️⃣ Reflection Transformation

Reflection mirrors vectors across an axis.

Reflection about x-axis:

$$
\begin{bmatrix}
1&0\\
0&-1
\end{bmatrix}
$$

```python
reflection = np.array([
    [1,0],
    [0,-1]
])

vector = np.array([3,4])

reflected = reflection @ vector

print("Original :", vector)
print("Reflected:", reflected)
```

# 6️⃣ Shear Transformation

Shearing slants the shape while preserving parallel lines.

Shear Matrix:

$$
\begin{bmatrix}
1&k\\
0&1
\end{bmatrix}
$$

```python
k = 1.5

shear = np.array([
    [1,k],
    [0,1]
])

vector = np.array([2,3])

sheared = shear @ vector

print("Original:", vector)
print("Sheared :", sheared)
```

# 7️⃣ Composition of Transformations

Multiple transformations can be combined.

If:

$$
T_1=A
$$

and

$$
T_2=B
$$

Then:

$$
T=B A
$$

This is called Composition.

```python
scale = np.array([
    [2,0],
    [0,2]
])

theta = np.radians(45)

rotation = np.array([
    [np.cos(theta),-np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

combined = rotation @ scale

print(combined)
```

# Applying Combined Transformation

We can apply multiple transformations in a single step.

```python
vector = np.array([1,1])

result = combined @ vector

print("Original :", vector)
print("Transformed:", result)
```

# 8️⃣ Transforming Point Clouds

A point cloud is a collection of points.

Transformations can be applied to all points simultaneously.

```python
points = np.array([
    [1,1],
    [2,1],
    [2,2],
    [1,2]
])

scale = np.array([
    [2,0],
    [0,2]
])

transformed_points = points @ scale.T

print(transformed_points)
```

# 🔟 Linear Transformations in NumPy

NumPy provides efficient matrix operations.

Common Functions:

- np.dot()
- @ operator
- np.matmul()

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print("Using @")
print(A @ B)

print("\nUsing np.matmul")
print(np.matmul(A,B))
```

# 🤖 Linear Transformations in Artificial Intelligence

Every Neural Network Layer computes:

$$
y = Wx+b
$$

where:

- W is a transformation matrix
- x is an input vector

The matrix multiplication:

$$
Wx
$$

is a Linear Transformation.

---

Applications:

### Machine Learning

Feature Transformations

### Computer Vision

Image Augmentation

### NLP

Word Embeddings

### Transformers

Query, Key, Value Projections

### Robotics

Coordinate Transformations

### Computer Graphics

3D Rendering Pipelines

```python
input_vector = np.array([
    1.0,
    2.0,
    3.0
])

weights = np.array([
    [0.2,0.5,0.1],
    [0.3,0.7,0.4]
])

output = weights @ input_vector

print("Input:")
print(input_vector)

print("\nOutput:")
print(output)
```
