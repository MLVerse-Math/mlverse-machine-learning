# Linear Transformations in Artificial Intelligence

## Neural Networks

A dense layer computes:

$$
y = Wx+b
$$

The matrix multiplication:

$$
Wx
$$

is a linear transformation.

---

## Computer Vision

Image transformations include:

- Rotation
- Scaling
- Cropping
- Reflection

used for data augmentation.

---

## Word Embeddings

Embedding matrices transform tokens into vector representations.

---

## Transformers

Attention layers perform multiple learned linear transformations.

Examples:

- Query Matrix
- Key Matrix
- Value Matrix

---

## Principal Component Analysis (PCA)

PCA transforms data into a new coordinate system.

---

## Robotics

Coordinate transformations convert sensor measurements between frames.

---

## Computer Graphics

3D rendering pipelines use linear transformations continuously.

---

## Large Language Models

LLMs apply billions of linear transformations during inference.
