# Mathematics of Linear Transformations

## General Transformation

$$
T(x)=Ax
$$

where:

$$
A=
\begin{bmatrix}
a&b\\
c&d
\end{bmatrix}
$$

and

$$
x=
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$

Result:

$$
T(x)=
\begin{bmatrix}
ax+by\\
cx+dy
\end{bmatrix}
$$

---

## Scaling

$$
S=
\begin{bmatrix}
s_x&0\\
0&s_y
\end{bmatrix}
$$

1&0\\

\end{bmatrix}

## Shear Transformation

$$
\begin{bmatrix}
1&k\\
0&1
\end{bmatrix}
$$

## Composition

If:

$$
T_1(x)=A_1x
$$

and

$$
T_2(x)=A_2x
$$

Then:

$$
T_2(T_1(x))
=
A_2A_1x
$$

Matrix multiplication combines transformations.
