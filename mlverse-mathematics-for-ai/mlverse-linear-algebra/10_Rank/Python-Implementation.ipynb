# What is Rank?

Rank is the number of linearly independent rows or columns in a matrix.

$$
Rank(A)
$$

It measures the amount of independent information contained in a matrix.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

print(A)
```

```python
rank = np.linalg.matrix_rank(A)

print("Rank =", rank)
```

# Full Rank Matrix

A matrix is full rank when:

$$
Rank(A)=\min(m,n)
$$

```python
A = np.array([
    [1, 2],
    [3, 4]
])

print(A)

print(
    "\nRank:",
    np.linalg.matrix_rank(A)
)
```

# Rank Deficient Matrix

Rows or columns contain redundant information.

```python
A = np.array([
    [1,2],
    [2,4]
])

print(A)

print(
    "\nRank:",
    np.linalg.matrix_rank(A)
)
```

# Row Dependency

The second row is a multiple of the first row.
$$
R_2 = 2R_1
$$

Therefore rank decreases.

```python
A = np.array([
    [1,2,3],
    [2,4,6],
    [3,6,9]
])

print(A)

print(
    "\nRank:",
    np.linalg.matrix_rank(A)
)
```

# Column Dependency

Columns can also be linearly dependent.

```python
A = np.array([
    [1,2],
    [2,4],
    [3,6]
])

print(A)

print(
    "\nRank:",
    np.linalg.matrix_rank(A)
)
```

# Rank and Determinants

If:

$$
det(A)\neq0
$$

then:

$$
Rank(A)=n
$$

```python
A = np.array([
    [2,3],
    [1,4]
])

det = np.linalg.det(A)

rank = np.linalg.matrix_rank(A)

print("Determinant =", det)

print("Rank =", rank)
```

# Singular Matrix

A singular matrix has:

$$
det(A)=0
$$

and reduced rank.

```python
A = np.array([
    [1,2],
    [2,4]
])

print(
    "Determinant:",
    np.linalg.det(A)
)

print(
    "Rank:",
    np.linalg.matrix_rank(A)
)
```

# Rank and Linear Systems

Consider:

$$
AX=B
$$

The rank determines whether a solution exists.

```python
A = np.array([
    [2,1],
    [1,-1]
])

B = np.array([
    5,
    1
])

solution = np.linalg.solve(A,B)

print(solution)
```

# Rank-Nullity Theorem

$$
Rank(A)+Nullity(A)=n
$$

where:

n = number of columns.

```python
A = np.array([
    [1,2,3],
    [2,4,6]
])

rank = np.linalg.matrix_rank(A)

n = A.shape[1]

nullity = n - rank

print("Rank =", rank)

print("Nullity =", nullity)
```

# Rank using Singular Value Decomposition (SVD)

SVD is one of the most powerful tools in AI and Machine Learning.

```python
A = np.array([
    [1,2],
    [3,4]
])

U,S,Vt = np.linalg.svd(A)

print("Singular Values:")

print(S)
```

```python
rank = np.sum(S > 1e-10)

print("Rank =", rank)
```

# Low Rank Approximation

Many AI systems compress information using low-rank approximations.

```python
A = np.random.rand(5,5)

U,S,Vt = np.linalg.svd(A)

k = 2

A_low_rank = (
    U[:,:k]
    @ np.diag(S[:k])
    @ Vt[:k,:]
)

print("Original Shape:", A.shape)

print(
    "Approximation Shape:",
    A_low_rank.shape
)
```

# Rank in Machine Learning Datasets

Rank tells us how many independent features exist.

```python
dataset = np.array([
    [1,2,3],
    [2,4,6],
    [3,6,9],
    [4,8,12]
])

print(dataset)

print(
    "\nFeature Rank:",
    np.linalg.matrix_rank(dataset)
)
```

# Rank in PCA

Principal Component Analysis identifies directions that preserve the maximum rank-related information.

PCA reduces dimensionality while retaining important structure.

```python
covariance_matrix = np.array([
    [4,2],
    [2,3]
])

eigenvalues, eigenvectors = np.linalg.eig(
    covariance_matrix
)

print("Eigenvalues:")

print(eigenvalues)
```

# Random Matrix Rank Analysis

```python
for size in [2,3,4,5]:

    A = np.random.randint(
        1,
        10,
        (size,size)
    )

    rank = np.linalg.matrix_rank(A)

    print("\nMatrix Size:", size)

    print(A)

    print("Rank =", rank)
```
