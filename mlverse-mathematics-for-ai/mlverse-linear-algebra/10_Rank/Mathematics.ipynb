# Mathematics of Rank

## Definition

Rank:

$$
Rank(A)
$$

is the dimension of the column space.

---

## Row Rank

Number of independent rows.

---

## Column Rank

Number of independent columns.

---

## Rank Theorem

For every matrix:

$$
Row\ Rank = Column\ Rank
$$

---

## Example

$$
A=
\begin{bmatrix}
1&2\\
2&4
\end{bmatrix}
$$

Since:

$$
Row_2 = 2Row_1
$$

Rank:

$$
Rank(A)=1
$$

---

## Full Rank

For:

$$
A=
\begin{bmatrix}
1&2\\
3&4
\end{bmatrix}
$$

$$
Rank(A)=2
$$

---

## Rank and Determinants

If:

$$
det(A)\neq0
$$

then:

$$
Rank(A)=n
$$

---

## Rank and Nullity

Rank-Nullity Theorem:

$$
Rank(A)+Nullity(A)=n
$$

where:

n = number of columns.

---

## Computational Complexity

Rank is usually computed using:

- Gaussian Elimination
- Singular Value Decomposition

Complexity:

$$
O(n^3)
$$
