# Rank Theory

## What is Rank?

Rank is the maximum number of linearly independent rows or columns in a matrix.

---

## Example

$$
A=
\begin{bmatrix}
1 & 2\\
2 & 4
\end{bmatrix}
$$

Second row is a multiple of first row.

Therefore:

$$
Rank(A)=1
$$

---

## Full Rank Matrix

A matrix is Full Rank when:

$$
Rank(A)=\min(m,n)
$$

Example:

$$
\begin{bmatrix}
1 & 2\\
3 & 4
\end{bmatrix}
$$

Rank = 2

---

## Rank Deficient Matrix

If:

$$
Rank(A)<\min(m,n)
$$

the matrix is rank deficient.

---

## Why Rank Matters

Rank tells us:

- Number of independent features
- Redundant information
- Solvability of systems
- Data dimensionality

---

## Rank and Linear Systems

For:

$$
AX=B
$$

Solution depends on matrix rank.

---

## Rank and Data

In Machine Learning:

Columns = Features

Rank tells how many truly independent features exist.
