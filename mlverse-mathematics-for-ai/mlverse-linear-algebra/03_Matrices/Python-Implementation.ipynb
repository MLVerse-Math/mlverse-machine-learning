```python
import numpy as np

print("NumPy Version:", np.__version__)
```

# Creating Matrices

A matrix is a rectangular arrangement of numbers organized into rows and columns.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

print(A)
```

# Matrix Dimensions

The shape of a matrix tells us:

(Number of Rows, Number of Columns)

```python
A = np.array([
    [1,2,3],
    [4,5,6]
])

print("Shape:", A.shape)
```

# Matrix Addition

Two matrices of the same dimensions can be added element-wise.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

C = A + B

print(C)
```

# Matrix Subtraction

Subtraction is performed element-wise.

```python
A = np.array([
    [5,6],
    [7,8]
])

B = np.array([
    [1,2],
    [3,4]
])

print(A - B)
```

# Scalar Multiplication

A scalar multiplies every element of a matrix.

```python
A = np.array([
    [1,2],
    [3,4]
])

print(3 * A)
```

# Matrix Multiplication

Matrix multiplication is one of the most important operations in AI.

Neural networks perform billions of matrix multiplications.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

C = A @ B

print(C)
```

# Matrix Transpose

Rows become columns.

```python
A = np.array([
    [1,2],
    [3,4]
])

print(A.T)
```

# Identity Matrix

An identity matrix behaves like the number 1 in matrix algebra.

```python
I = np.eye(3)

print(I)
```

# Determinant

The determinant helps determine whether a matrix is invertible.

```python
A = np.array([
    [1,2],
    [3,4]
])

det = np.linalg.det(A)

print(det)
```

# Matrix Inverse

The inverse matrix reverses a matrix transformation.

```python
A = np.array([
    [1,2],
    [3,4]
])

A_inv = np.linalg.inv(A)

print(A_inv)
```

# Matrix Rank

Rank tells us how many independent rows or columns exist.

```python
A = np.array([
    [1,2],
    [2,4]
])

rank = np.linalg.matrix_rank(A)

print(rank)
```

# Solving Linear Equations

Many Machine Learning problems can be expressed as systems of equations.

```python
A = np.array([
    [2,1],
    [1,3]
])

b = np.array([8,13])

x = np.linalg.solve(A,b)

print(x)
```

# Matrix Operations in Machine Learning

Feature Matrix:

Rows → Samples

Columns → Features

Example:

$$
X=
\begin{bmatrix}
Age & Salary & Experience
\end{bmatrix}
$$

Most machine learning algorithms operate directly on feature matrices.

```python
X = np.array([
    [25,50000,2],
    [30,70000,5],
    [35,90000,8]
])

print("Feature Matrix:")
print(X)

print("\nShape:", X.shape)
```
