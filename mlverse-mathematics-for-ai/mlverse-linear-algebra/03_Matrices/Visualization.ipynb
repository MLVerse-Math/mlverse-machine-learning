```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,6)
```

# Visualizing a Matrix

A matrix is a collection of numbers arranged in rows and columns.

```python
A = np.array([
    [1,2],
    [3,4]
])

plt.imshow(A)

plt.colorbar()

plt.title("Matrix Visualization")

plt.show()
```

# Matrix as a Transformation

Matrices can transform vectors.

We start with a vector:

$$
v=
\begin{bmatrix}
2\\
1
\end{bmatrix}
$$

```python
v = np.array([2,1])

plt.figure()

plt.quiver(
    0,0,
    v[0],
    v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.grid()

plt.title("Original Vector")

plt.show()
```

# Scaling Transformation

A scaling matrix changes the size of a vector.

```python
A = np.array([
    [2,0],
    [0,2]
])

v = np.array([2,1])

Av = A @ v

plt.figure()

plt.quiver(
    0,0,
    v[0],
    v[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Original'
)

plt.quiver(
    0,0,
    Av[0],
    Av[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Scaled'
)

plt.xlim(-10,10)
plt.ylim(-10,10)

plt.grid()

plt.legend()

plt.title("Scaling Transformation")

plt.show()
```

# Rotation Matrix

A rotation matrix rotates vectors.

$$
R=
\begin{bmatrix}
\cos\theta & -\sin\theta\\
\sin\theta & \cos\theta
\end{bmatrix}
$$

```python
theta = np.radians(45)

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta), np.cos(theta)]
])

v = np.array([3,1])

Rv = R @ v

plt.figure()

plt.quiver(
    0,0,
    v[0],
    v[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Original'
)

plt.quiver(
    0,0,
    Rv[0],
    Rv[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Rotated'
)

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.grid()

plt.legend()

plt.title("45 Degree Rotation")

plt.show()
```

# Shearing Transformation

Shearing shifts points in one direction while preserving parallel lines.

```python
S = np.array([
    [1,1],
    [0,1]
])

v = np.array([2,2])

Sv = S @ v

plt.figure()

plt.quiver(
    0,0,
    v[0],
    v[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Original'
)

plt.quiver(
    0,0,
    Sv[0],
    Sv[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Sheared'
)

plt.xlim(-6,6)
plt.ylim(-6,6)

plt.grid()

plt.legend()

plt.title("Shearing Matrix")

plt.show()
```

# Matrix Multiplication as Transformation

Matrix multiplication applies a transformation to a vector.

```python
A = np.array([
    [2,1],
    [1,2]
])

v = np.array([2,1])

Av = A @ v

plt.figure()

plt.quiver(
    0,0,
    v[0],
    v[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Original'
)

plt.quiver(
    0,0,
    Av[0],
    Av[1],
    angles='xy',
    scale_units='xy',
    scale=1,
    label='Transformed'
)

plt.xlim(-10,10)
plt.ylim(-10,10)

plt.grid()

plt.legend()

plt.title("Matrix Transformation")

plt.show()
```

# Visualizing the Unit Square

The unit square helps us understand matrix transformations geometrically.

```python
square = np.array([
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]
])

plt.figure()

plt.plot(
    square[:,0],
    square[:,1]
)

plt.grid()

plt.title("Unit Square")

plt.axis('equal')

plt.show()
```

# Transforming the Unit Square

Observe how the matrix changes the shape.

```python
A = np.array([
    [2,1],
    [1,2]
])

transformed = square @ A.T

plt.figure()

plt.plot(
    square[:,0],
    square[:,1],
    label="Original"
)

plt.plot(
    transformed[:,0],
    transformed[:,1],
    label="Transformed"
)

plt.grid()

plt.axis('equal')

plt.legend()

plt.title("Unit Square Transformation")

plt.show()
```

# Determinant as Area Scaling

The determinant tells us how much area changes after a transformation.

If:

$$
det(A)=2
$$

Area doubles.

If:

$$
det(A)=0
$$

The space collapses.

```python
A = np.array([
    [2,0],
    [0,3]
])

det = np.linalg.det(A)

print("Determinant =", det)
```
