# 🌌 Vector Spaces Visualization

In this notebook, we will visualize:

1. Vectors in R²
2. Linear Combinations
3. Span of Vectors
4. Subspaces
5. Basis Vectors
6. Dimension
7. Coordinate Systems
8. Feature Spaces
9. Embedding Spaces
10. Latent Spaces in AI

Visualization helps us understand how modern AI models learn and operate in vector spaces.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,6)
```

# 1️⃣ Visualizing R²

A vector space R² contains all vectors:

\[
(x,y)
\]

Each point in the plane is a vector.

```python
x = np.random.randint(-10,10,50)
y = np.random.randint(-10,10,50)

plt.scatter(x,y)

plt.axhline(0)
plt.axvline(0)

plt.grid(True)

plt.title("Vectors in R²")

plt.show()
```

# 2️⃣ Linear Combination Visualization

Given:

$$
v_1=(1,0)
$$

$$
v_2=(0,1)
$$

Any vector in R² can be generated as:

$$
av_1+bv_2
$$

```python
v1 = np.array([1,0])
v2 = np.array([0,1])

a = 3
b = 4

result = a*v1 + b*v2

plt.figure()

plt.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1)
plt.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1)
plt.quiver(0,0,result[0],result[1],angles='xy',scale_units='xy',scale=1)

plt.xlim(-1,6)
plt.ylim(-1,6)

plt.grid()

plt.title("Linear Combination")

plt.show()
```

# 3️⃣ Span of Two Vectors

The span represents all vectors that can be generated using linear combinations.

```python
v1 = np.array([1,0])
v2 = np.array([0,1])

points = []

for a in range(-5,6):
    
    for b in range(-5,6):
        
        p = a*v1 + b*v2
        
        points.append(p)

points = np.array(points)

plt.scatter(
    points[:,0],
    points[:,1]
)

plt.grid()

plt.title("Span(v₁,v₂)")

plt.show()
```

# 4️⃣ Subspace Visualization

The x-axis is a subspace of R².

All vectors satisfy:

$$
y=0
$$

```python
x = np.linspace(-10,10,100)

y = np.zeros_like(x)

plt.plot(x,y)

plt.grid()

plt.title("Subspace of R²")

plt.show()
```

# 5️⃣ Basis Visualization

The standard basis vectors of R² are:

$$
e_1=(1,0)
$$

$$
e_2=(0,1)
$$

```python
e1 = np.array([1,0])
e2 = np.array([0,1])

plt.figure()

plt.quiver(
    0,0,
    e1[0],e1[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    e2[0],e2[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(-1,2)
plt.ylim(-1,2)

plt.grid()

plt.title("Basis Vectors")

plt.show()
```

# 6️⃣ Dimension Visualization

Dimension represents the number of basis vectors required to describe a space.

Examples:

R² → Dimension 2

R³ → Dimension 3

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

ax = fig.add_subplot(
    111,
    projection='3d'
)

ax.quiver(
    0,0,0,
    1,0,0
)

ax.quiver(
    0,0,0,
    0,1,0
)

ax.quiver(
    0,0,0,
    0,0,1
)

ax.set_title("Dimension of R³")

plt.show()
```

# 7️⃣ Coordinate Representation

Any vector can be represented using basis vectors.

```python
v = np.array([4,3])

plt.figure()

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()

plt.title("Coordinate Representation")

plt.show()
```

# 8️⃣ Feature Space Visualization

Machine Learning data points exist in feature spaces.

```python
age = np.random.randint(
    18,
    60,
    100
)

salary = np.random.randint(
    20000,
    120000,
    100
)

plt.scatter(
    age,
    salary
)

plt.xlabel("Age")
plt.ylabel("Salary")

plt.title("Feature Space")

plt.grid()

plt.show()
```

# 9️⃣ Embedding Space Visualization

Word embeddings are vectors in high-dimensional spaces.

For visualization, we project them into 2D.

```python
embeddings = np.random.randn(
    20,
    2
)

plt.scatter(
    embeddings[:,0],
    embeddings[:,1]
)

plt.title(
    "Embedding Space"
)

plt.grid()

plt.show()
```

# 🔟 Latent Space Visualization

Generative AI models learn latent spaces.

Nearby points often represent similar concepts.

```python
latent = np.random.randn(
    100,
    2
)

plt.scatter(
    latent[:,0],
    latent[:,1]
)

plt.title(
    "Latent Space"
)

plt.grid()

plt.show()
```
