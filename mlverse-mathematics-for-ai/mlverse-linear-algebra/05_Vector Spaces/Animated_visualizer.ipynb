# 🎬 Interactive Linear Combination

Move the sliders to change the coefficients:
$$

v = a v_1 + b v_2

$$

Observe how the resulting vector moves inside the vector space.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, FloatSlider

v1 = np.array([1,0])
v2 = np.array([0,1])

def linear_combination(a,b):

    result = a*v1 + b*v2

    plt.figure(figsize=(7,7))

    plt.quiver(
        0,0,
        v1[0],v1[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label='v1'
    )

    plt.quiver(
        0,0,
        v2[0],v2[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label='v2'
    )

    plt.quiver(
        0,0,
        result[0],result[1],
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.grid(True)

    plt.title(
        f"Result = {a:.1f}v₁ + {b:.1f}v₂"
    )

    plt.show()

interact(
    linear_combination,
    a=FloatSlider(-5, min=-10,max=10,step=0.5),
    b=FloatSlider(3, min=-10,max=10,step=0.5)
);
```

# 🎬 Span Generator

This animation generates all possible vectors inside the span.

Increasing the range expands the visible span.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, IntSlider

v1 = np.array([1,0])
v2 = np.array([0,1])

def span_visualizer(limit):

    points = []

    for a in range(-limit,limit+1):

        for b in range(-limit,limit+1):

            points.append(
                a*v1+b*v2
            )

    points = np.array(points)

    plt.figure(figsize=(7,7))

    plt.scatter(
        points[:,0],
        points[:,1]
    )

    plt.grid(True)

    plt.title(
        f"Span Size = {limit}"
    )

    plt.show()

interact(
    span_visualizer,
    limit=IntSlider(
        value=3,
        min=1,
        max=20
    )
);
```

# 🎬 Basis Vector Construction

Watch how basis vectors combine to create any vector.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact

def basis_builder(x,y):

    plt.figure(figsize=(7,7))

    plt.quiver(
        0,0,
        1,0,
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        x,0,
        0,y,
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        0,0,
        x,y,
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.grid(True)

    plt.title(
        f"Vector ({x},{y})"
    )

    plt.show()

interact(
    basis_builder,
    x=(-8,8,1),
    y=(-8,8,1)
);
```

# 🎬 Machine Learning Feature Space

Each point represents a sample.

Adjust the number of samples to see how feature spaces grow.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact

def feature_space(samples):

    age = np.random.randint(
        18,
        60,
        samples
    )

    salary = np.random.randint(
        20000,
        120000,
        samples
    )

    plt.figure(figsize=(8,6))

    plt.scatter(
        age,
        salary
    )

    plt.xlabel("Age")
    plt.ylabel("Salary")

    plt.grid(True)

    plt.title(
        f"{samples} Samples"
    )

    plt.show()

interact(
    feature_space,
    samples=(10,1000,10)
);
```

# 🎬 Word Embedding Space

Move the slider to generate embeddings.

This simulates how words occupy regions of semantic space.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact

def embedding_space(n):

    embeddings = np.random.randn(
        n,
        2
    )

    plt.figure(figsize=(8,8))

    plt.scatter(
        embeddings[:,0],
        embeddings[:,1]
    )

    plt.grid(True)

    plt.title(
        f"{n} Embeddings"
    )

    plt.show()

interact(
    embedding_space,
    n=(10,500,10)
);
```

# 🎬 Latent Space Navigation

Generative AI models move through latent spaces.

Use the slider to move through the space.

```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact

latent = np.random.randn(
    200,
    2
)

def latent_navigation(index):

    plt.figure(figsize=(8,8))

    plt.scatter(
        latent[:,0],
        latent[:,1],
        alpha=0.3
    )

    point = latent[index]

    plt.scatter(
        point[0],
        point[1],
        s=200
    )

    plt.grid(True)

    plt.title(
        f"Latent Point {index}"
    )

    plt.show()

interact(
    latent_navigation,
    index=(0,199,1)
);
```
