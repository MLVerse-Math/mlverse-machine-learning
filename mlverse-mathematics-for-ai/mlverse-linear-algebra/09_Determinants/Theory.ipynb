# Determinants Theory

## What is a Determinant?

A determinant is a numerical value calculated from a square matrix.

Example:

\[
A=
\begin{bmatrix}
2 & 3\\
1 & 4
\end{bmatrix}
\]

The determinant is:

\[
det(A)
\]

---

## Why Are Determinants Important?

Determinants tell us:

- Whether a matrix is invertible
- Whether equations have unique solutions
- How areas and volumes change
- Whether vectors are linearly independent

---

## Geometric Meaning

For a 2×2 matrix:

$$
A=
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
$$

The determinant represents the area scaling factor.

---

## Singular Matrix

If:

$$
det(A)=0
$$

The matrix is singular.

It cannot be inverted.

---

## Non-Singular Matrix

If:

$$
det(A)\neq0
$$

The matrix is invertible.

---

## Determinants and Linear Systems

For:

$$
AX=B
$$

A unique solution exists only if:

$$
det(A)\neq0
$$

---

## Determinants and Transformations

Determinants measure how much a transformation stretches or shrinks space.
