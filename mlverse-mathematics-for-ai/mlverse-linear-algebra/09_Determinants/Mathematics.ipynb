# Mathematics of Determinants

## 2×2 Determinant

Given:

$$
A=
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
$$

The determinant is:

$$
det(A)=ad-bc
$$

---

## Example

$$
A=
\begin{bmatrix}
2 & 3\\
1 & 4
\end{bmatrix}
$$

$$
det(A)=2(4)-3(1)
$$

$$
det(A)=5
$$

---

## 3×3 Determinant

$$
A=
\begin{bmatrix}
a & b & c\\
d & e & f\\
g & h & i
\end{bmatrix}
$$

$$
det(A)=
a(ei-fh)
-b(di-fg)
+c(dh-eg)
$$

---

## Minors

The minor of an element is the determinant of the submatrix obtained after removing its row and column.

---

## Cofactors

$$
C_{ij}=(-1)^{i+j}M_{ij}
$$

---

## Laplace Expansion

$$
det(A)=
\sum a_{ij}C_{ij}
$$

---

## Properties

### Identity Matrix

$$
det(I)=1
$$

### Transpose

$$
det(A^T)=det(A)
$$

### Product

$$
det(AB)=det(A)det(B)
$$

### Inverse

$$
det(A^{-1})
=
\frac1{det(A)}
$$

---

## Computational Complexity

Efficient algorithms:

$$
O(n^3)
$$

using LU decomposition.
