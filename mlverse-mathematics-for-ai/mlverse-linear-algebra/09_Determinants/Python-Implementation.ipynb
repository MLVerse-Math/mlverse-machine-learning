# Determinant of a 2×2 Matrix

Given:

$$
A=
\begin{bmatrix}
a & b\\
c & d
\end{bmatrix}
$$

Determinant:

$$
det(A)=ad-bc
$$

```python
A = np.array([
    [2,3],
    [1,4]
])

print(A)
```

```python
det = np.linalg.det(A)

print("Determinant =", det)
```

# Computing Determinant Manually

```python
a,b = 2,3
c,d = 1,4

det = a*d - b*c

print(det)
```

# Determinant of a 3×3 Matrix

```python
A = np.array([
    [1,2,3],
    [4,5,6],
    [7,8,10]
])

print(A)
```

```python
det = np.linalg.det(A)

print("Determinant =", det)
```

# Cofactor Expansion

For a 3×3 matrix:

$$
det(A)=
a(ei−fh)
-b(di−fg)
+c(dh−eg)
$$

```python
A = np.array([
    [1,2,3],
    [4,5,6],
    [7,8,10]
])

a,b,c = A[0]

d,e,f = A[1]

g,h,i = A[2]

det = (
    a*(e*i - f*h)
    - b*(d*i - f*g)
    + c*(d*h - e*g)
)

print(det)
```

# Singular Matrix

A matrix is singular when:

$$
det(A)=0
$$
Such matrices cannot be inverted.

```python
A = np.array([
    [1,2],
    [2,4]
])

print(A)

print(
    "Determinant:",
    np.linalg.det(A)
)
```

# Non-Singular Matrix

A matrix is invertible when:

$$
det(A)\neq0
$$

```python
A = np.array([
    [2,3],
    [1,4]
])

print(
    "Determinant:",
    np.linalg.det(A)
)

print(
    "\nInverse Matrix:\n",
    np.linalg.inv(A)
)
```

# Determinants and Matrix Inverses

The determinant determines whether an inverse exists.

```python
matrices = [

    np.array([
        [2,3],
        [1,4]
    ]),

    np.array([
        [1,2],
        [2,4]
    ])
]

for i,A in enumerate(matrices):

    print(f"\nMatrix {i+1}")

    det = np.linalg.det(A)

    print("Determinant:", det)

    if abs(det) > 1e-10:
        print("Invertible")

    else:
        print("Not Invertible")
```

# Determinant of Random Matrices

```python
for size in [2,3,4,5]:

    A = np.random.randint(
        1,
        10,
        (size,size)
    )

    det = np.linalg.det(A)

    print("\nSize:", size)

    print(A)

    print(
        "Determinant:",
        round(det,2)
    )
```

# Determinants and Area Scaling

Consider:

$$
A=
\begin{bmatrix}
2 & 0\\
0 & 3
\end{bmatrix}
$$

Area scales by:

$$
det(A)=6
$$

```python
A = np.array([
    [2,0],
    [0,3]
])

print(A)

print(
    "Scaling Factor:",
    np.linalg.det(A)
)
```

# Determinants in Machine Learning

Principal Component Analysis (PCA)

Requires solving:

$$
det(A-\lambda I)=0
$$

to find eigenvalues.

```python
covariance_matrix = np.array([
    [4,2],
    [2,3]
])

print(
    "Covariance Matrix:\n",
    covariance_matrix
)

print(
    "\nDeterminant:",
    np.linalg.det(
        covariance_matrix
    )
)
```

# Characteristic Equation

Used in Eigenvalue Computation

$$
det(A-\lambda I)=0
$$

```python
A = np.array([
    [4,2],
    [1,3]
])

eigenvalues = np.linalg.eigvals(A)

print(
    "Eigenvalues:"
)

print(
    eigenvalues
)
```

# Performance Benchmark

Determinant computation becomes expensive for large matrices.

```python
import time

sizes = [10,50,100]

for n in sizes:

    A = np.random.rand(n,n)

    start = time.time()

    np.linalg.det(A)

    end = time.time()

    print(
        f"{n}x{n} Matrix:",
        round(end-start,4),
        "seconds"
    )
```
