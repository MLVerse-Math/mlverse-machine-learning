
# Mathematics of Orthogonality

## Dot Product

For vectors:

$$
u=
\begin{bmatrix}
u_1\\
u_2
\end{bmatrix}
$$

$$
v=
\begin{bmatrix}
v_1\\
v_2
\end{bmatrix}
$$

Dot Product:

$$
u \cdot v
=
u_1v_1+u_2v_2
$$
