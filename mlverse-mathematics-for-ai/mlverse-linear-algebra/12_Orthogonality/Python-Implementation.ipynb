# Dot Product

For vectors:

$$
u=[u_1,u_2]
$$

$$
v=[v_1,v_2]
$$

Dot Product:

$$
u^Tv
$$

```python
u = np.array([1, 2])

v = np.array([3, 4])

dot_product = np.dot(u, v)

print("u =", u)
print("v =", v)

print("\nDot Product =", dot_product)
```

# Orthogonal Vectors

Two vectors are orthogonal is:
$$
u^Tv=0
$$


```python
u = np.array([1, 0])

v = np.array([0, 1])

dot_product = np.dot(u, v)

print("Dot Product =", dot_product)

if dot_product == 0:
    print("Vectors are Orthogonal")
else:
    print("Vectors are NOT Orthogonal")
```

# Non-Orthogonal Example

# Vector Norm

Magnitude of a vector:

$$
||u|| =\sqrt{x^2+y^2}
$$

```python
u = np.array([3, 4])

norm = np.linalg.norm(u)

print("Norm =", norm)
```

# Unit Vector

A vector with magnitude 1.

```python
u = np.array([3, 4])

unit_u = u / np.linalg.norm(u)

print(unit_u)

print(
    "Norm:",
    np.linalg.norm(unit_u)
)
```

# Orthonormal Vectors

Orthogonal + Unit Length

```python
u = np.array([1,0])

v = np.array([0,1])

print(
    "Dot Product:",
    np.dot(u,v)
)

print(
    "Norm u:",
    np.linalg.norm(u)
)

print(
    "Norm v:",
    np.linalg.norm(v)
)
```

# Angle Between Vectors

$$
\cos(\theta)
=
\frac{u^Tv}
{||u|| ||v||}
$$

```python
u = np.array([1,2])

v = np.array([3,1])

cos_theta = (
    np.dot(u,v)
    /
    (
        np.linalg.norm(u)
        *
        np.linalg.norm(v)
    )
)

theta = np.degrees(
    np.arccos(cos_theta)
)

print(
    "Angle:",
    theta
)
```

# Orthogonal Basis

A basis where all vectors are orthogonal.

```python
basis = np.array([
    [1,0],
    [0,1]
])

for i in range(len(basis)):
    for j in range(i+1,len(basis)):
        print(
            np.dot(
                basis[i],
                basis[j]
            )
        )
```

# Orthogonal Matrix

A matrix is orthogonal if:
$$
Q^TQ=I
$$

```python
Q = np.array([
    [0,-1],
    [1, 0]
])

result = Q.T @ Q

print(result)
```

```python
print(
    np.allclose(
        result,
        np.eye(2)
    )
)
```

# Vector Projection

Projection of x onto u:

$$
Proj_u(x)
=
\frac{x^Tu}{u^Tu}u
$$

```python
x = np.array([4,3])

u = np.array([1,0])

projection = (
    np.dot(x,u)
    /
    np.dot(u,u)
) * u

print(projection)
```

# Gram-Schmidt Orthogonalization

Convert independent vectors into orthogonal vectors.

```python
v1 = np.array(
    [1,1]
)

v2 = np.array(
    [1,0]
)

u1 = v1

u2 = (
    v2
    -
    (
        np.dot(v2,u1)
        /
        np.dot(u1,u1)
    ) * u1
)

print("u1 =", u1)

print("u2 =", u2)

print(
    "Dot Product:",
    np.dot(u1,u2)
)
```

# Normalized Gram-Schmidt

```python
e1 = u1 / np.linalg.norm(u1)

e2 = u2 / np.linalg.norm(u2)

print(e1)

print(e2)

print(
    np.dot(e1,e2)
)
```

# QR Decomposition

$$
A=QR
$$

Q = Orthogonal Matrix

R = Upper Triangular Matrix

```python
A = np.array([
    [1,2],
    [3,4]
])

Q,R = np.linalg.qr(A)

print("Q =\n")
print(Q)

print("\nR =\n")
print(R)
```

```python
print(
    "\nQᵀQ =\n"
)

print(
    Q.T @ Q
)
```

# PCA Orthogonality

Principal Components are orthogonal directions.

```python
from sklearn.decomposition import PCA

data = np.random.randn(
    500,
    5
)

pca = PCA(
    n_components=5
)

pca.fit(data)

components = pca.components_

print(
    components.shape
)
```

```python
orthogonality_check = (
    components
    @
    components.T
)

print(
    np.round(
        orthogonality_check,
        4
    )
)
```

# Orthogonal Initialization

Used in Deep Learning for stable training.

```python
random_matrix = np.random.randn(
    5,
    5
)

Q,R = np.linalg.qr(
    random_matrix
)

weights = Q

print(
    "Orthogonal Weight Matrix:"
)

print(weights)
```

# High-Dimensional Orthogonality

Random vectors become nearly orthogonal in high dimensions.

```python
v1 = np.random.randn(1000)

v2 = np.random.randn(1000)

cos_similarity = (
    np.dot(v1,v2)
    /
    (
        np.linalg.norm(v1)
        *
        np.linalg.norm(v2)
    )
)

print(
    "Cosine Similarity:",
    cos_similarity
)
```
