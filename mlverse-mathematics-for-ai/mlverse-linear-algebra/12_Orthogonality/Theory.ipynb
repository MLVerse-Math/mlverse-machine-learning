# Orthogonality Theory

## What is Orthogonality?

Two vectors are orthogonal if they are perpendicular.

Example:

$$
u=
\begin{bmatrix}
1\\
0
\end{bmatrix}
$$

$$
v=
\begin{bmatrix}
0\\
1
\end{bmatrix}
$$

These vectors are orthogonal.

---

## Geometric Interpretation

Orthogonal vectors form a 90° angle.

They represent completely independent directions.

---

## Why Orthogonality Matters

Orthogonal vectors:

- Reduce redundancy
- Simplify computations
- Improve numerical stability
- Form efficient coordinate systems

---

## Orthogonal Basis

A basis where every pair of vectors is orthogonal.

---

## Orthonormal Basis

Orthogonal basis with unit length vectors.

---

## Real World Meaning

Orthogonality means information independence.

In Machine Learning this often translates to:

- Less redundancy
- Better feature representation
- Better dimensionality reduction
