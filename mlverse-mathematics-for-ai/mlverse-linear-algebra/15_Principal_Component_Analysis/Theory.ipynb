# Principal Component Analysis Theory

## What is PCA?

Principal Component Analysis (PCA) is a dimensionality reduction technique.

Instead of using original features, PCA creates new features called Principal Components.

---

## Principal Components

A Principal Component is a direction in feature space that captures maximum variance.

First Principal Component:

- Captures maximum variance.

Second Principal Component:

- Captures maximum remaining variance.
- Orthogonal to the first component.

---

## Why PCA Works

Many datasets contain correlated features.

PCA transforms them into uncorrelated components.

---

## Core Idea

Find directions where data varies the most.

These directions contain the most information.

---

## Result

Transform:

$$
n \text{ Features}
$$

into:

$$
k \text{ Principal Components}
$$

where:

$$
k < n
$$

while preserving most information.
