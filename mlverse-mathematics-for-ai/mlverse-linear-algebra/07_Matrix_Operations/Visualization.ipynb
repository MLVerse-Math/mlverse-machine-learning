# 1️⃣ Matrix Addition Visualization

Observe how two vectors combine through addition.

Matrix addition can be understood as combining transformations.

```python
def vector_addition(x1,y1,x2,y2):

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,x1,y1,
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        0,0,x2,y2,
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.quiver(
        0,0,
        x1+x2,
        y1+y2,
        angles='xy',
        scale_units='xy',
        scale=1
    )

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.grid(True)

    plt.title("Vector Addition")

    plt.show()

interact(
    vector_addition,
    x1=(-5,5,0.5),
    y1=(-5,5,0.5),
    x2=(-5,5,0.5),
    y2=(-5,5,0.5)
);
```

# 2️⃣ Scalar Multiplication Explorer

A scalar scales every element of a matrix.

Move the slider to stretch or compress the shape.

```python
square = np.array([
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]
])

def scalar_scale(k):

    transformed = square * k

    plt.figure(figsize=(8,8))

    plt.plot(
        square[:,0],
        square[:,1],
        linewidth=3,
        label="Original"
    )

    plt.plot(
        transformed[:,0],
        transformed[:,1],
        linewidth=3,
        label="Scaled"
    )

    plt.grid(True)

    plt.legend()

    plt.title(
        f"Scalar = {k:.2f}"
    )

    plt.show()

interact(
    scalar_scale,
    k=(-5,5,0.1)
);
```

# 3️⃣ Matrix Multiplication Transformation

Matrix multiplication creates geometric transformations.

Adjust matrix values and observe the transformation.

```python
shape = np.array([
    [0,0],
    [1,0],
    [1,1],
    [0,1],
    [0,0]
])

def matrix_transform(a,b,c,d):

    A = np.array([
        [a,b],
        [c,d]
    ])

    transformed = shape @ A.T

    plt.figure(figsize=(8,8))

    plt.plot(
        shape[:,0],
        shape[:,1],
        linewidth=3,
        label="Original"
    )

    plt.plot(
        transformed[:,0],
        transformed[:,1],
        linewidth=3,
        label="Transformed"
    )

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.grid(True)

    plt.legend()

    plt.title(
        f"Matrix\n{A}"
    )

    plt.show()

interact(
    matrix_transform,
    a=(-3,3,0.1),
    b=(-3,3,0.1),
    c=(-3,3,0.1),
    d=(-3,3,0.1)
);
```

# 4️⃣ Determinant Visualizer

The determinant measures area scaling.

Observe how the area changes as the determinant changes.

```python
def determinant_visualizer(k):

    A = np.array([
        [k,0],
        [0,1]
    ])

    shape = np.array([
        [0,0],
        [1,0],
        [1,1],
        [0,1],
        [0,0]
    ])

    transformed = shape @ A.T

    determinant = np.linalg.det(A)

    plt.figure(figsize=(8,8))

    plt.fill(
        transformed[:,0],
        transformed[:,1],
        alpha=0.3
    )

    plt.plot(
        transformed[:,0],
        transformed[:,1]
    )

    plt.grid(True)

    plt.title(
        f"Determinant = {determinant:.2f}"
    )

    plt.xlim(-6,6)
    plt.ylim(-6,6)

    plt.show()

interact(
    determinant_visualizer,
    k=(-5,5,0.1)
);
```

# 5️⃣ Matrix Inverse Explorer

A matrix inverse reverses a transformation.

Apply a transformation and then its inverse.

```python
def inverse_demo(scale):

    A = np.array([
        [scale,0],
        [0,scale]
    ])

    inverse = np.linalg.inv(A)

    point = np.array([2,1])

    transformed = A @ point

    recovered = inverse @ transformed

    print("Original :", point)
    print("Transformed :", transformed)
    print("Recovered :", recovered)

interact(
    inverse_demo,
    scale=(0.5,5,0.1)
);
```

# 6️⃣ Rank Visualization

Rank measures the number of independent directions.

If rows become dependent, rank decreases.

```python
def rank_visualizer(k):

    A = np.array([
        [1,2],
        [k,2*k]
    ])

    rank = np.linalg.matrix_rank(A)

    print("Matrix:\n")
    print(A)

    print("\nRank =", rank)

interact(
    rank_visualizer,
    k=(-5,5,0.1)
);
```

# 7️⃣ Interactive Matrix Playground

Experiment with any 2×2 matrix and see its effect on the coordinate grid.

This is one of the best ways to understand linear algebra visually.

```python
def grid_transform(a,b,c,d):

    A = np.array([
        [a,b],
        [c,d]
    ])

    plt.figure(figsize=(9,9))

    for x in range(-5,6):

        line = np.array([
            [x,-5],
            [x,5]
        ])

        transformed = line @ A.T

        plt.plot(
            transformed[:,0],
            transformed[:,1]
        )

    for y in range(-5,6):

        line = np.array([
            [-5,y],
            [5,y]
        ])

        transformed = line @ A.T

        plt.plot(
            transformed[:,0],
            transformed[:,1]
        )

    plt.xlim(-15,15)
    plt.ylim(-15,15)

    plt.grid(True)

    plt.title(
        f"Transformation Matrix\n{A}"
    )

    plt.show()

interact(
    grid_transform,
    a=(-3,3,0.1),
    b=(-3,3,0.1),
    c=(-3,3,0.1),
    d=(-3,3,0.1)
);
```

# 8️⃣ Neural Network Matrix Computation

Every neural network layer performs matrix multiplication.

$$
y = Wx + b
$$

This visualization demonstrates a simple neural layer.

```python
input_vector = np.array([
    1,
    2,
    3
])

weights = np.array([
    [0.5,0.2,0.1],
    [0.3,0.7,0.4]
])

bias = np.array([
    0.1,
    0.2
])

output = weights @ input_vector + bias

print("Input Vector:")
print(input_vector)

print("\nWeight Matrix:")
print(weights)

print("\nBias:")
print(bias)

print("\nOutput:")
print(output)
```
