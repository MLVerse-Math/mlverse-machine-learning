# 1️⃣ Creating Matrices

A matrix is a rectangular arrangement of numbers.

Example:

$$
A=
\begin{bmatrix}
1 & 2\\
3 & 4
\end{bmatrix}
$$

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)
```

# 2️⃣ Matrix Addition

Matrices with the same dimensions can be added element-wise.

```python
result = A + B

print("A + B =")
print(result)
```

# 3️⃣ Matrix Subtraction

Subtract corresponding elements from two matrices.

```python
result = B - A

print("B - A =")
print(result)
```

# 4️⃣ Scalar Multiplication

Multiply every element of a matrix by a scalar value.

```python
scalar = 3

result = scalar * A

print(result)
```

# 5️⃣ Matrix Multiplication

Matrix multiplication combines rows of one matrix with columns of another matrix.

This is one of the most important operations in AI.

```python
result = A @ B

print("A × B =")
print(result)
```

# Understanding Dimensions

If:

$$
A_{m \times n}
$$

and

$$
B_{n \times p}
$$

Then:
$$
AB_{m \times p}
$$

```python
print("Shape of A:", A.shape)
print("Shape of B:", B.shape)

print("\nResult Shape:")
print((A @ B).shape)
```

# 6️⃣ Element-wise Multiplication

Also called the Hadamard Product.

Every element is multiplied by its corresponding element.

```python
result = A * B

print(result)
```

# 7️⃣ Matrix Transpose

Transpose converts rows into columns.

```python
A_transpose = A.T

print("A:")
print(A)

print("\nAᵀ:")
print(A_transpose)
```

# 8️⃣ Identity Matrix

The identity matrix is the matrix equivalent of the number 1.

$$
AI=A
$$

```python
I = np.eye(3)

print(I)
```

```python
A = np.array([
    [2,3],
    [1,4]
])

I = np.eye(2)

print(A @ I)
```

# 9️⃣ Matrix Determinant

The determinant measures how much a matrix scales space.

For a 2×2 matrix:

$$
det(A)=ad-bc
$$

```python
A = np.array([
    [2,3],
    [1,4]
])

determinant = np.linalg.det(A)

print("Determinant:")
print(determinant)
```

# 🔟 Matrix Inverse

The inverse reverses a matrix transformation.

$$
AA^{-1}=I
$$

```python
A = np.array([
    [2,3],
    [1,4]
])

inverse = np.linalg.inv(A)

print("Inverse Matrix:")
print(inverse)
```

```python
print("Verification:")

print(
    np.round(
        A @ inverse,
        decimals=5
    )
)
```

# 1️⃣1️⃣ Matrix Rank

Rank represents the number of independent rows or columns.

```python
A = np.array([
    [1,2],
    [2,4]
])

rank = np.linalg.matrix_rank(A)

print("Rank:", rank)
```

```python
A = np.array([
    [1,2],
    [3,4]
])

print(
    np.linalg.matrix_rank(A)
)
```

# 1️⃣2️⃣ Matrix Trace

The trace is the sum of diagonal elements.

$$
tr(A)
=
a_{11}+a_{22}+...
$$

```python
A = np.array([
    [1,2],
    [3,4]
])

trace = np.trace(A)

print(trace)
```

# 1️⃣3️⃣ Useful NumPy Matrix Functions

NumPy provides many optimized matrix operations.

```python
A = np.array([
    [1,2],
    [3,4]
])

print("Transpose:")
print(A.T)

print("\nDeterminant:")
print(np.linalg.det(A))

print("\nInverse:")
print(np.linalg.inv(A))

print("\nRank:")
print(np.linalg.matrix_rank(A))

print("\nTrace:")
print(np.trace(A))
```

# 1️⃣4️⃣ Matrix Operations in Neural Networks

A neural network layer computes:

$$
y = Wx+b
$$

where:

- W = Weight Matrix
- x = Input Vector
- b = Bias Vector

This operation is essentially matrix multiplication.

```python
input_vector = np.array([
    1.0,
    2.0,
    3.0
])

weights = np.array([
    [0.5,0.2,0.1],
    [0.4,0.8,0.3]
])

bias = np.array([
    0.1,
    0.2
])

output = weights @ input_vector + bias

print("Input:")
print(input_vector)

print("\nOutput:")
print(output)
```

# Mini Neural Network Example

Input → Matrix Multiplication → Output

```python
X = np.array([
    [1,2],
    [3,4],
    [5,6]
])

W = np.array([
    [0.5,0.2],
    [0.8,0.4]
])

Y = X @ W

print(Y)
```

# Matrix Operations Behind Transformers

Transformers use matrix multiplication extensively.

Examples:

$$
QK^T
$$

$$
Attention = Softmax(QK^T)V
$$

Every token is represented as a vector and transformed through matrices.

```python
Q = np.random.rand(4,3)
K = np.random.rand(4,3)

attention_scores = Q @ K.T

print(attention_scores)
```
