# Systems of Linear Equations Theory

## Introduction

A linear equation is an equation of degree one.

Example:

$$
2x+3y=7
$$

A system consists of multiple such equations.

---

## Example

$$
\begin{cases}
2x+y=5 \\
x-y=1
\end{cases}
$$

---

## Types of Systems

### Consistent System

Has at least one solution.

---

### Inconsistent System

No solution exists.

Example:

$$
x+y=2
$$

$$
x+y=5
$$

---

### Dependent System

Infinite solutions.

Example:

$$
x+y=2
$$

$$
2x+2y=4
$$

---

## Methods of Solution

### Substitution

Solve one variable and substitute.

### Elimination

Remove variables systematically.

### Matrix Method

Represent equations as matrices.

### Gaussian Elimination

Convert matrix into Row Echelon Form.

### Matrix Inverse Method

Use inverse matrix.

### Cramer's Rule

Use determinants.

---

## Why Important?

Many AI algorithms solve systems of equations internally:

- Linear Regression
- Neural Networks
- Optimization
- Kalman Filters
- Robotics
