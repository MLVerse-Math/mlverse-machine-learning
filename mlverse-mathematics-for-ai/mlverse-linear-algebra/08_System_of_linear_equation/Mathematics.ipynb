# Mathematics of Systems of Linear Equations

## Standard Form

$$
a_1x+b_1y=c_1
$$

$$
a_2x+b_2y=c_2
$$

---

## Matrix Form

$$
AX=B
$$

Where

$$
A=
\begin{bmatrix}
a_1 & b_1\\
a_2 & b_2
\end{bmatrix}
$$

$$
X=
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$

$$
B=
\begin{bmatrix}
c_1\\
c_2
\end{bmatrix}
$$

---

## Matrix Solution

$$
X=A^{-1}B
$$

---

## Example

$$
\begin{bmatrix}
2 & 1\\
1 & -1
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
\begin{bmatrix}
5\\
1
\end{bmatrix}
$$

Solution:

$$
x=2
$$

$$
y=1
$$

---

## Gaussian Elimination

Augmented Matrix:

$$
\left[
\begin{array}{cc|c}
2&1&5\\
1&-1&1
\end{array}
\right]
$$

Perform row operations until Row Echelon Form.

---

## Determinant Method

For

$$
AX=B
$$

$$
x=\frac{D_x}{D}
$$

$$
y=\frac{D_y}{D}
$$

where

$$
D=det(A)
$$

---

## Computational Complexity

Gaussian Elimination:

$$
O(n^3)
$$

Widely used in scientific computing and AI.
