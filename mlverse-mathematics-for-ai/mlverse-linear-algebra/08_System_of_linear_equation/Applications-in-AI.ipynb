# Applications of Systems of Linear Equations in AI

Systems of equations appear everywhere in AI.

---

## Linear Regression

Normal Equation:

$$
(X^TX)\theta=X^Ty
$$

Solve for:

$$
\theta
$$

---

## Neural Networks

Backpropagation creates systems of equations.

Weight updates depend on solving gradients.

---

- Camera Calibration
- 3D Reconstruction
- Image Alignment

---

## Robotics

Used in:

- Kinematics
- Localization
- Motion Planning

## Kalman Filters

State estimation uses matrix equations.

Most optimization algorithms solve linear systems repeatedly.
