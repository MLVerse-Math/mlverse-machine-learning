# Example System

$$
\begin{cases}
2x+y=5 \\
x-y=1
\end{cases}
$$

Matrix Form:

$$
AX=B
$$

$$
A=
\begin{bmatrix}
2 & 1 \\
1 & -1
\end{bmatrix}
$$

$$
B=
\begin{bmatrix}
5 \\
1
\end{bmatrix}
$$

```python
A = np.array([
    [2, 1],
    [1, -1]
])

B = np.array([
    5,
    1
])

print("A =\n", A)
print("\nB =\n", B)
```

# Solving Using NumPy

NumPy provides the `linalg.solve()` function.

```python
solution = np.linalg.solve(A, B)

print("Solution:")
print("x =", solution[0])
print("y =", solution[1])
```

# Matrix Inverse Method

The solution is:

$$
X=A^{-1}B
$$

```python
A_inv = np.linalg.inv(A)

solution = A_inv @ B

print("Inverse Matrix:\n")
print(A_inv)

print("\nSolution:")
print(solution)
```

# Determinant

A matrix is invertible only if:

$$
det(A)\neq0
$$

```python
determinant = np.linalg.det(A)

print("Determinant =", determinant)
```

# Gaussian Elimination From Scratch

We solve the system using row operations.

```python
def gaussian_elimination(A, B):

    A = A.astype(float)
    B = B.astype(float)

    n = len(B)

    for i in range(n):

        pivot = A[i][i]

        A[i] = A[i] / pivot
        B[i] = B[i] / pivot

        for j in range(i + 1, n):

            factor = A[j][i]

            A[j] -= factor * A[i]
            B[j] -= factor * B[i]

    x = np.zeros(n)

    for i in range(n - 1, -1, -1):

        x[i] = B[i]

        for j in range(i + 1, n):
            x[i] -= A[i][j] * x[j]

    return x
```

```python
A = np.array([
    [2,1],
    [1,-1]
])

B = np.array([
    5,
    1
])

solution = gaussian_elimination(A, B)

print("Solution:")
print(solution)
```

# Cramer's Rule

For:

$$
AX=B
$$

$$
x=\frac{D_x}{D}
$$

$$
y=\frac{D_y}{D}
$$

```python
A = np.array([
    [2,1],
    [1,-1]
])

B = np.array([
    5,
    1
])

D = np.linalg.det(A)

Dx = np.linalg.det(np.array([
    [5,1],
    [1,-1]
]))

Dy = np.linalg.det(np.array([
    [2,5],
    [1,1]
]))

x = Dx / D
y = Dy / D

print("x =", x)
print("y =", y)
```

# Solving a 3×3 System

Real-world AI systems often involve thousands or millions of equations.

```python
A = np.array([
    [1,1,1],
    [2,-1,1],
    [1,2,-1]
])

B = np.array([
    6,
    3,
    3
])

solution = np.linalg.solve(A, B)

print(solution)
```

# Interactive System Solver

Adjust coefficients and instantly solve the system.

```python
from ipywidgets import interact
from ipywidgets import FloatSlider
```

```python
def solve_system(a,b,c,d,e,f):

    A = np.array([
        [a,b],
        [d,e]
    ])

    B = np.array([
        c,
        f
    ])

    try:

        solution = np.linalg.solve(A,B)

        print("x =", round(solution[0],4))
        print("y =", round(solution[1],4))

    except:

        print("No unique solution")
```

```python
interact(
    solve_system,

    a=FloatSlider(min=-10,max=10,value=2),
    b=FloatSlider(min=-10,max=10,value=1),
    c=FloatSlider(min=-20,max=20,value=5),

    d=FloatSlider(min=-10,max=10,value=1),
    e=FloatSlider(min=-10,max=10,value=-1),
    f=FloatSlider(min=-20,max=20,value=1)
);
```

# AI Application

## Linear Regression Normal Equation

One of the most famous systems of equations in Machine Learning:

$$
(X^TX)\theta = X^Ty
$$

Solving this equation gives the optimal parameters for linear regression.

```python
X = np.array([
    [1,1],
    [1,2],
    [1,3],
    [1,4]
])

y = np.array([
    2,
    3,
    4,
    5
])

theta = np.linalg.inv(
    X.T @ X
) @ (X.T @ y)

print("Parameters:")
print(theta)
```
