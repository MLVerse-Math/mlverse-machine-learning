# Basis and Dimension Theory

## What is a Basis?

A basis is a set of linearly independent vectors that span a vector space.

Every vector in the space can be represented as a combination of basis vectors.

---

## Example

In 2D:

$$
\mathbf{e_1}
=
\begin{bmatrix}
1\\
0
\end{bmatrix}
$$

$$
\mathbf{e_2}
=
\begin{bmatrix}
0\\
1
\end{bmatrix}
$$

These vectors form a basis for:

$$
\mathbb{R}^2
$$

---

## Why Basis Matters

Basis vectors provide a coordinate system.

Any vector can be expressed as:

$$
v = c_1e_1 + c_2e_2
$$

---

## What is Dimension?

Dimension is the number of vectors in a basis.

Examples:

$$
dim(\mathbb{R}^2)=2
$$

$$
dim(\mathbb{R}^3)=3
$$

---

## Linear Independence

Basis vectors must be independent.

No basis vector can be created from others.

---

## Span

Basis vectors must span the entire space.

---

## Minimal Representation

A basis provides the smallest set of vectors needed to represent all vectors in a space.
