# Mathematics of Basis and Dimension

## Basis Definition

A set:

$$
B=
\{v_1,v_2,\ldots,v_n\}
$$

is a basis if:

1. Linearly Independent
2. Spans the Space

---

## Standard Basis of R²

$$
e_1=
\begin{bmatrix}
1\\
0
\end{bmatrix}
$$

$$
e_2=
\begin{bmatrix}
0\\
1
\end{bmatrix}
$$

---

## Standard Basis of R³

$$
e_1=
\begin{bmatrix}
1\\
0\\
0
\end{bmatrix}
$$

$$
e_2=
\begin{bmatrix}
0\\
1\\
0
\end{bmatrix}
$$

$$
e_3=
\begin{bmatrix}
0\\
0\\
1
\end{bmatrix}
$$

---

## Coordinate Representation

For basis:

$$
B=\{e_1,e_2\}
$$

Vector:

$$
v=
\begin{bmatrix}
3\\
2
\end{bmatrix}
$$

can be written as:

$$
v=3e_1+2e_2
$$

---

## Dimension

$$
dim(V)=|B|
$$

where:

B = Basis

---

## Basis and Rank

$$
dim(ColumnSpace)=Rank(A)
$$

---

## Rank-Dimension Relationship

Rank determines the number of basis vectors in the column space.
