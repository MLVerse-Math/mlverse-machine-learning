# Basis and Dimension in Artificial Intelligence

Basis and Dimension are everywhere in AI.

---

## Feature Spaces

Features act as dimensions.

Example:

Age
Income
Education

Each feature creates a new dimension.

---

## Word Embeddings

Words are represented in high-dimensional vector spaces.

Example:
$$
Word \rightarrow \mathbb{R}^{768}
$$

for BERT embeddings.

---

## Principal Component Analysis

PCA creates a new basis.

The new basis captures maximum variance.

---

## Computer Vision

Images are represented in extremely high-dimensional spaces.

Basis vectors help compress information.

---

## Deep Learning

Hidden layers learn useful basis representations.

---

## Latent Spaces

Generative models learn compressed basis representations.

Examples:

- Autoencoders
- Variational Autoencoders
- Diffusion Models

---

## Recommendation Systems

Matrix factorization discovers latent basis vectors.

---

## Large Language Models

Transformer embeddings operate in learned vector spaces.
