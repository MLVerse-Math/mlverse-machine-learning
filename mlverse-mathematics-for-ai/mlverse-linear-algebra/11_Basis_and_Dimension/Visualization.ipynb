# Standard Basis Visualization

Visualize the standard basis vectors of R².

```python
def standard_basis():

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,
        1,0,
        color='red',
        scale=1,
        scale_units='xy'
    )

    plt.quiver(
        0,0,
        0,1,
        color='lime',
        scale=1,
        scale_units='xy'
    )

    plt.text(1.1,0,'e1',fontsize=14,color='red')
    plt.text(0,1.1,'e2',fontsize=14,color='lime')

    plt.grid(True)

    plt.xlim(-2,2)
    plt.ylim(-2,2)

    plt.title("Standard Basis")

    plt.show()

standard_basis()
```

# Span Explorer

Observe how basis vectors span the plane.

```python
def span_visualizer(a,b):

    e1 = np.array([a,0])
    e2 = np.array([0,b])

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,
        e1[0],e1[1],
        color='cyan',
        scale=1,
        scale_units='xy'
    )

    plt.quiver(
        0,0,
        e2[0],e2[1],
        color='magenta',
        scale=1,
        scale_units='xy'
    )

    for i in range(-5,6):
        for j in range(-5,6):

            p = i*e1 + j*e2

            plt.scatter(
                p[0],
                p[1],
                color='yellow',
                alpha=0.4
            )

    plt.grid(True)

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.title("Span of Basis Vectors")

    plt.show()

interact(
    span_visualizer,
    a=FloatSlider(min=0.5,max=3,value=1),
    b=FloatSlider(min=0.5,max=3,value=1)
);
```

# Linear Independence Explorer

```python
def independence(a,b,c,d):

    A = np.array([
        [a,b],
        [c,d]
    ])

    rank = np.linalg.matrix_rank(A)

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,
        a,b,
        color='red',
        scale=1,
        scale_units='xy'
    )

    plt.quiver(
        0,0,
        c,d,
        color='blue',
        scale=1,
        scale_units='xy'
    )

    if rank == 2:
        title = "Independent Vectors"
    else:
        title = "Dependent Vectors"

    plt.title(title)

    plt.grid(True)

    plt.xlim(-10,10)
    plt.ylim(-10,10)

    plt.show()
```

```python
interact(
    independence,

    a=FloatSlider(min=-5,max=5,value=1),
    b=FloatSlider(min=-5,max=5,value=0),

    c=FloatSlider(min=-5,max=5,value=0),
    d=FloatSlider(min=-5,max=5,value=1)
);
```

# Basis Transformation Visualizer

Observe coordinate system changes.

```python
def transform(a,b,c,d):

    basis = np.array([
        [a,b],
        [c,d]
    ])

    e1 = basis[:,0]
    e2 = basis[:,1]

    plt.figure(figsize=(8,8))

    plt.quiver(
        0,0,
        1,0,
        color='white',
        scale=1,
        scale_units='xy'
    )

    plt.quiver(
        0,0,
        0,1,
        color='white',
        scale=1,
        scale_units='xy'
    )

    plt.quiver(
        0,0,
        e1[0],
        e1[1],
        color='red',
        scale=1,
        scale_units='xy'
    )

    plt.quiver(
        0,0,
        e2[0],
        e2[1],
        color='cyan',
        scale=1,
        scale_units='xy'
    )

    plt.grid(True)

    plt.xlim(-6,6)
    plt.ylim(-6,6)

    plt.title("Basis Transformation")

    plt.show()
```

```python
interact(
    transform,

    a=FloatSlider(min=-3,max=3,value=1),
    b=FloatSlider(min=-3,max=3,value=1),

    c=FloatSlider(min=-3,max=3,value=1),
    d=FloatSlider(min=-3,max=3,value=-1)
);
```
