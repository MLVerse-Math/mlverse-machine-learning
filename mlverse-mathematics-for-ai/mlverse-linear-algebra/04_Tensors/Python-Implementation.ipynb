# Creating Tensors

A tensor is a multidimensional array.

Examples:

- Scalar → Rank 0
- Vector → Rank 1
- Matrix → Rank 2
- Higher Dimensional Arrays → Rank 3+

```python
scalar = np.array(5)

vector = np.array([1,2,3])

matrix = np.array([
    [1,2],
    [3,4]
])

tensor3d = np.array([
    [
        [1,2],
        [3,4]
    ],
    [
        [5,6],
        [7,8]
    ]
])

print("Scalar:\n", scalar)
print("\nVector:\n", vector)
print("\nMatrix:\n", matrix)
print("\nTensor:\n", tensor3d)
```

# Tensor Shapes

The shape tells us the size of every dimension.

```python
tensor = np.array([
    [
        [1,2],
        [3,4]
    ],
    [
        [5,6],
        [7,8]
    ]
])

print("Shape:", tensor.shape)
```

# Tensor Rank

Rank represents the number of dimensions.

```python
scalar = np.array(10)

vector = np.array([1,2,3])

matrix = np.array([
    [1,2],
    [3,4]
])

tensor = np.random.rand(2,3,4)

print("Scalar Rank:", scalar.ndim)
print("Vector Rank:", vector.ndim)
print("Matrix Rank:", matrix.ndim)
print("Tensor Rank:", tensor.ndim)
```

# Tensor Reshaping

Reshaping changes dimensions without changing data.

```python
tensor = np.arange(12)

print("Original:")
print(tensor)

reshaped = tensor.reshape(3,4)

print("\nReshaped:")
print(reshaped)
```

# Tensor Transpose

Transpose swaps dimensions.

```python
matrix = np.array([
    [1,2,3],
    [4,5,6]
])

print("Original:")
print(matrix)

print("\nTranspose:")
print(matrix.T)
```

# Tensor Addition

Addition is performed element-wise.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

C = A + B

print(C)
```

# Tensor Broadcasting

Broadcasting automatically expands dimensions.

```python
A = np.array([
    [1,2,3],
    [4,5,6]
])

B = np.array([10,20,30])

print(A + B)
```

# Tensor Multiplication

Element-wise multiplication.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print(A * B)
```

# Matrix Multiplication

Different from element-wise multiplication.

```python
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print(A @ B)
```

# Tensor Slicing

Access specific portions of a tensor.

```python
tensor = np.array([
    [
        [1,2],
        [3,4]
    ],
    [
        [5,6],
        [7,8]
    ]
])

print(tensor[0])

print("\n")

print(tensor[:,:,1])
```

# Tensor Statistics

Useful for Machine Learning preprocessing.

```python
tensor = np.random.rand(5,5)

print("Mean:", np.mean(tensor))
print("Max :", np.max(tensor))
print("Min :", np.min(tensor))
print("Std :", np.std(tensor))
```

# Tensor Operations with NumPy

NumPy is the foundation of scientific computing.

```python
tensor = np.random.rand(3,3)

print("Tensor:")
print(tensor)

print("\nSum:")
print(np.sum(tensor))

print("\nMean:")
print(np.mean(tensor))

print("\nShape:")
print(tensor.shape)
```

# Tensor Operations with PyTorch

PyTorch tensors power modern Deep Learning.

```python
import torch

tensor = torch.tensor([
    [1,2],
    [3,4]
])

print(tensor)

print("\nShape:")
print(tensor.shape)

print("\nRank:")
print(tensor.ndim)
```

```python
A = torch.tensor([
    [1,2],
    [3,4]
])

B = torch.tensor([
    [5,6],
    [7,8]
])

print(A + B)

print("\n")

print(torch.matmul(A,B))
```

# GPU Acceleration with PyTorch

PyTorch can move tensors to the GPU.

```python
import torch

device = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

print("Device:", device)

tensor = torch.rand(
    (3,3),
    device=device
)

print(tensor)
```
