````python
import numpy as np
import plotly.graph_objects as go

from ipywidgets import (
    FloatSlider,
    VBox,
    interactive_output
)

from IPython.display import display
````

# What is a Vector?

A vector is an ordered collection of numbers.

Example:

$$
v=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

The vector has:

- X Component = 3
- Y Component = 4

Geometrically, a vector can be visualized as an arrow.

````python
def vector_visualizer(x,y):

    magnitude = np.sqrt(x**2+y**2)

    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x=[0,x],
            y=[0,y],
            mode="lines+markers",
            name="Vector"
        )
    )

    fig.update_layout(
        title=f"Vector [{x:.1f},{y:.1f}] | Magnitude = {magnitude:.2f}",
        width=700,
        height=600,
        xaxis=dict(range=[-10,10]),
        yaxis=dict(range=[-10,10])
    )

    fig.show()

x_slider = FloatSlider(
    value=3,
    min=-10,
    max=10,
    step=0.1,
    description="X",
    continuous_update=False
)

y_slider = FloatSlider(
    value=4,
    min=-10,
    max=10,
    step=0.1,
    description="Y",
    continuous_update=False
)

display(VBox([x_slider,y_slider]))

display(
    interactive_output(
        vector_visualizer,
        {"x":x_slider,"y":y_slider}
    )
)
````

# Vector Magnitude

The magnitude (length) of a vector is:

$$
||v||
=
\sqrt{x^2+y^2}
$$

For:

$$
v=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

The magnitude becomes:

$$
5
$$

This is the famous Pythagorean Theorem.

````python
def magnitude_explorer(x,y):

    magnitude = np.linalg.norm([x,y])

    print(
        f"Magnitude = {magnitude:.4f}"
    )

display(
    interactive_output(
        magnitude_explorer,
        {
            "x":x_slider,
            "y":y_slider
        }
    )
)
````

# Vector Addition

Given:

$$
u=
\begin{bmatrix}
u_1\\
u_2
\end{bmatrix}
$$

and
$$
v=
\begin{bmatrix}
v_1\\
v_2
\end{bmatrix}
$$

Then:

$$
u+v
=
\begin{bmatrix}
u_1+v_1\\
u_2+v_2
\end{bmatrix}
$$

Vector addition combines displacement.

````python
def vector_addition(x1,y1,x2,y2):

    u=np.array([x1,y1])
    v=np.array([x2,y2])

    result=u+v

    fig=go.Figure()

    fig.add_trace(
        go.Scatter(
            x=[0,u[0]],
            y=[0,u[1]],
            mode='lines+markers',
            name='u'
        )
    )

    fig.add_trace(
        go.Scatter(
            x=[0,v[0]],
            y=[0,v[1]],
            mode='lines+markers',
            name='v'
        )
    )

    fig.add_trace(
        go.Scatter(
            x=[0,result[0]],
            y=[0,result[1]],
            mode='lines+markers',
            name='u+v'
        )
    )

    fig.update_layout(
        width=700,
        height=600,
        title=f"u+v={result}"
    )

    fig.show()
````

````python
display(
    VBox([
        FloatSlider(value=2,min=-10,max=10,description='x1'),
        FloatSlider(value=3,min=-10,max=10,description='y1'),
        FloatSlider(value=4,min=-10,max=10,description='x2'),
        FloatSlider(value=1,min=-10,max=10,description='y2')
    ])
)
````

# Scalar Multiplication

A scalar changes the size of a vector.

$$
kv
=
\begin{bmatrix}
kx\\
ky
\end{bmatrix}
$$

If:

$$
k>1
$$

the vector grows.

If:

$$
0<k<1
$$

the vector shrinks.

If:

$$
k<0
$$

the direction reverses.

````python
def scalar_multiplication(x,y,k):

    v=np.array([x,y])

    scaled=k*v

    fig=go.Figure()

    fig.add_trace(
        go.Scatter(
            x=[0,v[0]],
            y=[0,v[1]],
            mode='lines+markers',
            name='Original'
        )
    )

    fig.add_trace(
        go.Scatter(
            x=[0,scaled[0]],
            y=[0,scaled[1]],
            mode='lines+markers',
            name='Scaled'
        )
    )

    fig.show()
````

````python
scalar_multiplication(2,4,1)
````

# Dot Product

The dot product measures similarity.

$$
u \cdot v
=
u_1v_1+u_2v_2
$$

Applications:

- Search Engines
- Recommendation Systems
- Embeddings
- Machine Learning

````python
def dot_product(u1,u2,v1,v2):

    u=np.array([u1,u2])

    v=np.array([v1,v2])

    print(
        "Dot Product:",
        np.dot(u,v)
    )
````

````python
dot_product([1,2],[2,3],[1,0],[0,1])
````

# Angle Between Vectors

The angle between vectors is:

$$
\cos\theta
=
\frac{u \cdot v}
{||u||\,||v||}
$$

This formula powers:

- Semantic Search
- Vector Databases
- Recommendation Systems
- LLM Embeddings

````python
def angle_explorer(u1,u2,v1,v2):

    u=np.array([u1,u2])

    v=np.array([v1,v2])

    angle=np.degrees(
        np.arccos(
            np.dot(u,v)/
            (
                np.linalg.norm(u)
                *
                np.linalg.norm(v)
            )
        )
    )

    print(
        f"Angle = {angle:.2f}°"
    )
````

````python
angle_explorer(1,2,2,2)
````
